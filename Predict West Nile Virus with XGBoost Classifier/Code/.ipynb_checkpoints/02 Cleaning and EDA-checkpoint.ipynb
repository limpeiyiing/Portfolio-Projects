## Custom Functions

```python
def check_df(df):
    #check for df shape, null, and duplicate
    print('Dataset Statistics:')
    print(f'Shape of dataframe: {df.shape}')
    print('--------------------------------------')
    print(f'Null values in dataframe: {df.isna().sum().sum()}')
    print('--------------------------------------')
    print(f'% of Null values in dataframe: {round(((df.isna().sum().sum())/(df.shape[0])) * 100, 2)}%')
    print('--------------------------------------')
    print(f"Total duplicate rows: {df[df.duplicated()].shape[0]}")
    print(f"% duplicate rows: {round(df[df.duplicated()].shape[0] / df.shape[0] * 100, 2)}%")
    print('--------------------------------------')
    print(f'Column names: {df.columns}')
    print('--------------------------------------')
    print('\nVariable Types')
    print(f"Columns Count: \n{df.dtypes.value_counts()}")
```

```python
#custom function for data housekeeping
def data_housekeep(dfs):
    for df in dfs:
        #changing all to lower case
        df.columns = df.columns.str.lower()
        #converting Date into datetime
        df.date = pd.to_datetime(df.date)         
```

```python
#custom fuction for adding month, year and week features
def add_periods(df,date_col):
    df['year'] = df[date_col].dt.year
    df['month'] = df[date_col].dt.month
    df['week']= df['date'].dt.isocalendar().week
```

```python
#custom function to find distance between coordinates
def distance_finder(coordinate_1, coordinate_2):
    return np.sqrt((coordinate_1[0]-coordinate_2[0])**2 + (coordinate_1[1]-coordinate_2[1])**2)
```

```python
#function to get cluster of trap nearest to the centroids
def nearest_trap(trap_coord):
    dist = np.exp(100)
    group = group_centroids.index
    for i in range(4):
        current_distance = distance_finder(trap_coord, tuple(group_centroids.iloc[i,:]))
        if (current_distance < dist):
            trap_group = group[i]
            dist = current_distance
    return trap_group
```

```python
#custom function for hist plot
def hist_plot(x,y,label):
    palette = ['tab:blue', 'tab:orange']
    plt.figure(figsize=(15,5))
    sns.histplot(data=weather, x=x,y=y, hue="station", kde = True, palette=palette)
    plt.title(f"Distribution of {label} by Station",fontsize="x-large",fontweight="heavy");
    return
```

```python
def scatter_plotter(df, year, color, marker, size, ax):
    lats = df[df['year'] == year]['latitude']
    longs = df[df['year'] == year]['longitude']
    ax.scatter(x=longs, y=lats, c=color, s=size, marker=marker)
    
def map_plot_trap(year, ax):
    origin = [41.6, -88.0]              # lat/long of origin (lower left corner)
    upperRight = [42.1, -87.5]          # lat/long of upper right corner

    mapdata = np.loadtxt("../dataset/input/mapdata_copyright_openstreetmap_contributors.txt")

    intersection = [41.909614, -87.746134]  # co-ordinates of intersection of IL64 / IL50 according to Google Earth

    # generate plot
    ax.imshow(mapdata, cmap=plt.get_cmap('gray'), extent=[origin[1], upperRight[1], origin[0], upperRight[0]])

    # points showing west nile virus presence
    scatter_plotter(train[train['wnvpresent']==0], year, 'purple', 'x', 15, ax)
    scatter_plotter(train[train['wnvpresent']==1], year, 'b', 'x', 20, ax)
    
    weather_station_1 = ax.scatter(-87.933, 41.995, c="red", marker='v', s=50)
    weather_station_2 = ax.scatter(-87.752, 41.786, c="orange", marker='v', s=50)
    
    ax.set_title('Trap and Location of West Nile Virus in {}'.format(year))
    ax.legend(['Trap with No WNV', 'Trap with WNV', 'Station 1', 'Station 2'], fontsize=12);
```

#### Clean Up data

Features to impute: tavg, wetbulb, preciptotal

TAVG: Using the average of tmax & tmin to impute into the missing tavg

```python
#impute the missing
for i in weather.loc[weather.tavg=='M'].index:
    weather.loc[i,'tavg'] = round((weather.loc[i,'tmax']+weather.loc[i,'tmin'])/2,0)
#convert the column dtype from object to float
weather.tavg = weather.tavg.astype(float)
```

```python
avg_temp = weather.pivot_table(values=['tavg'],
                                     index=['date'],
                                     columns='station',
                                     aggfunc='mean')
```

```python
avg_temp.plot(kind="line", stacked=False,figsize=(15,5) );
plt.title('Preciptotal camparision of the two weather station')
plt.legend(['station 1', 'Station 2'], fontsize=12);
```

<b>Findings:</b>
- The temperature between the two stations are quite similar.
- Thus we will be using the distance between the station and trap to merge up the train and weather sets.

WETBULB: has some missing values. Impute with (temperature minus dewpoint). Then take this number and divide by 3. Subtract this number from the temperature.([source](https://theweatherprediction.com/habyhints/170/#:~:text=A%20quick%20technique%20that%20many,this%20number%20from%20the%20temperature)):

```python
#impute the missing
for i in weather.loc[weather['wetbulb']== 'M'].index:
    weather.loc[i,'wetbulb'] = weather.loc[i,'tavg'] - \
    ((weather.loc[i,'tavg'] - weather.loc[i,'dewpoint'])/3)
#convert the column dtype from object to float
weather.wetbulb = weather.wetbulb.astype(int)
```

##### Relative Humidity

In general, mosquito is more active when humidity is high so we want to take a look at it. ([source](https://www.orkin.com/other/mosquitoes/activity#:~:text=Warm%20temperatures%20of%20approximately%2080,and%20may%20even%20cause%20mortality.)).

Formula of Relative Humidity is given as ([source](https://www.omnicalculator.com/physics/relative-humidity)):

$\ln RH=\frac{\frac{}{}\frac{T_d{}}{b}(a-\frac{aT}{b+T}) - \frac{aT}{b+T}}{\frac{T_d{}}{b} + 1}$

where $a$ = 17.27, $b$ = 237.7, $T$ = Temperature in Celcius, $T_d$ = Dew Point Temperature in Celcius

```python
weather['humidity'] = 100*(np.exp((17.67 * ((weather['dewpoint'] - 32) * (5/9))) /
                                           (243.5 + ((weather['dewpoint'] - 32) * (5/9))))
                                    /
                                    np.exp((17.67 * ((weather['tavg'] - 32) * (5/9))) /
                                           (243.5 + ((weather['tavg'] - 32) * (5/9)))))
```

```python
humid = weather.pivot_table(values=['humidity'],
                                     index=['year','week'],
                                     columns='station',
                                     aggfunc='mean')
```

```python
humid.plot(kind= 'line',figsize=(15,5));
plt.title('Humidity camparision of the two weather station')
plt.legend(['station 1', 'Station 2'], fontsize=12);
```

<b>Findings:</b>
- Same as the preciptotal, the humidity do have some difference between the two station.
- the values are quite similar.
- Thus we will also be using the distance between the station and trap to merge up the train and weather sets.

```python
# coordinates for both stations
coord_s1 = (41.995, -87.933) 
coord_s2 = (41.786, -87.752) 

from geopy import distance
# for each row compare distance from each station using geopy
for i in range(len(train)):
    coord = (train['latitude'][i],train['longitude'][i])
    dist_s1 = abs(distance.distance(coord_s1,coord).km)
    dist_s2 = abs(distance.distance(coord_s2,coord).km)
    # creating new column for distance from respective Station
    train.loc[i,'distfrom_s1']=round(dist_s1,1)
    train.loc[i,'distfrom_s2']=round(dist_s2,1)
    
    # station that is closest to the trap will be assigned accordingly
    if dist_s1<dist_s2:
        train.loc[i,'station']=1
    elif dist_s2<dist_s1:
        train.loc[i,'station']=2
```

```python
trap_location = train_combined.groupby('trap')[['latitude', 'longitude']].max()
trap_stat = train_combined.groupby('trap')[['nummosquitos','wnvpresent']].sum()
traps_set = trap_location.merge(trap_stat, left_index=True, right_index=True)

# As clustering is done by K-Means framework, all features need to be scaled first
ss = StandardScaler()
traps_s = ss.fit_transform(traps_set)

# KMeans is deployed to cluster the traps
kmeans = KMeans(n_clusters=5)
kmeans.fit(traps_s)

# Result of clustering is kept in DataFrame format
trap_group = pd.DataFrame(data = kmeans.predict(traps_s), columns=['trap_group'], index=traps_set.index)
traps_set = traps_set.merge(trap_group, left_index=True, right_index=True)
```

#### Lag & Rolling Average

Weather data is provided in day to day form as such we need to smoothen the data.

##### 3 day shift and rolling

We did a 3 day lag to cater that the mosquitoes took some time to breed

```python
columns_to_smoothen = ['tavg', 'humidity', 'preciptotal']
new_column_names = []
for columns in columns_to_smoothen:
    new_column_names.append(columns + '_lag3_rol3')
    
train_combined[new_column_names] = train_combined[columns_to_smoothen]\
                                     .shift(3).rolling(3, center=True)\
                                     .mean().fillna(method='ffill')\
                                     .fillna(method='bfill')
```
