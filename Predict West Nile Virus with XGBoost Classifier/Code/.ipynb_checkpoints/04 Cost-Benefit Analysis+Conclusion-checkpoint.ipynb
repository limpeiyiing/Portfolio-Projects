# Cost-Benefit Analysis (Part 4/4)

----

### The strategy:
1) Identify the best time period to apply vector control. Spraying insecticides too early, before the season starts is a waste of resources because it will do little to minimize infections. Conversely, spraying insecticides too late will save vector control costs, but increase the risk of infections and therefore healthcare costs. 

2) Identify hotspots in the city to apply vector control to best control mosquito populations and therefore incidence of WNV.


------------

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
spray = pd.read_csv('../dataset/spray.csv')
train = pd.read_csv('../dataset/train.csv')
weather = pd.read_csv('../dataset/weather_cleaned.csv')
combined = pd.read_csv('../dataset/train_combined_latest.csv')
```

## Identifying the best time period and duration to apply vector control

```python
train.columns = train.columns.str.lower()
train['date'] = pd.to_datetime(train['date'])
train_agg = train.groupby(['date','latitude', 'longitude','species','trap'])[['date', 
                                                                              'species', 
                                                                              'trap', 
                                                                              'nummosquitos',
                                                                              'wnvpresent']].sum(['nummosquitos','row_count']).reset_index(drop=False)
train_agg['year'] = train_agg['date'].dt.year
train_agg['month'] = train_agg['date'].dt.month
train_agg['coordinates']=[(lat,long) for lat,long in zip(train_agg['latitude'],train_agg['longitude'])]
train_agg['wnvpresent'] = train_agg['wnvpresent'].map(lambda x:1 if x>0 else 0)
```

```python
## Aggregating the number of mosquitos and presence of WNV per trap by monthly period:
trap_by_period = train_agg.groupby(["trap","year","month"])[['nummosquitos','wnvpresent']].sum(['nummosquitos','wnvpresent']).reset_index(drop=False)
trap_by_period['wnvpresent'] = trap_by_period['wnvpresent'].map(lambda x:1 if x>0 else 0)
```

## Monthly incidences of WNV and Sum of Mosquitos 2007-2013:

From monthly plots of incidence of WNV by trap counts, we can see a seasonal trend of WNV appearing in July every year, and peaking in August. Based on this, we know that we could start regular surveillance and vector control methods in July instead of May or June. 

```python
year_list = [2007,2009,2011,2013]

fig, axes = plt.subplots(nrows=1, ncols=4, sharex=False, 
                             sharey=True, squeeze=True, figsize=(18,3), 
                             constrained_layout=True)
axes = axes.ravel() 
for i, year in enumerate(year_list): 
    tp = trap_by_period.loc[trap_by_period["year"]==year,:]
    rows = tp['month']
    cols = tp['wnvpresent']
    trap_wnv = pd.crosstab(rows,cols)
    trap_wnv.plot(kind='bar', stacked=True,ax = axes[i])
    axes[i].set_title(f'Trap Count by Month in {year}',fontsize=15,fontweight="heavy");
```

```python
rows = trap_by_period['month']
cols = trap_by_period['year']
vals = trap_by_period['nummosquitos']
trap_mos = pd.crosstab(rows,cols,vals,aggfunc='sum').fillna(0)

fig, axes = plt.subplots(nrows=1, ncols=4, sharex=False, 
                             sharey=True, squeeze=True, figsize=(18,3), 
                             constrained_layout=True)
axes = axes.ravel() 
for i, year in enumerate(year_list): 
    trap_mos[year].plot(kind='bar',ax = axes[i])
    axes[i].set_title(f'Sum of Mosquitos by Month in {year}',fontsize=15,fontweight="heavy");
```

## Weekly incidences of WNV and Sum of Mosquitos 2007-2013:

#### Transforming and merging data for visualisation purposes

```python
#aggregating to remove duplicated lines due to station 
weather['date'] = pd.to_datetime(weather['date'])
weather['year'] = weather['date'].dt.year
weather['month'] = weather['date'].dt.month
weather_agg = weather.groupby("date")[["date","tavg","preciptotal","humidity","year","month"]].mean(["tavg","preciptotal","humidity"]).reset_index(drop=False)

weather_agg.set_index("date",drop=True,inplace=True)

#Use weekly rolling mean to smoothen the line plot
weather_subset = weather_agg.merge(weather_agg[['tavg','preciptotal', 'humidity']].rolling(7).mean(),
                left_index=True, right_index=True,how="inner",
                                 suffixes = ('_0','_rolling_week_average')).dropna()
weather_subset.head()
```

```python
# Create an aggregated view of number of mosquitos and incidence of wnv by trap count in year-week buckets
trap_by_date = train_agg.groupby(["trap","date",'year'])[['nummosquitos','wnvpresent']].sum(['nummosquitos','wnvpresent']).reset_index(drop=False)
trap_by_date['wnvpresent'] = trap_by_date['wnvpresent'].map(lambda x:1 if x>0 else 0)
trap_by_date['trap_count'] = 1 
trap_by_date["week"] = trap_by_date["date"].dt.isocalendar().week
trap_by_date.head()
```

```python
# create a master reference of closing date per week based on weather dataset 
# weather dataset has the most complete date list 

weather["date"] = pd.to_datetime(weather["date"])
weather["year"] = weather["date"].dt.year
weather["month"] = weather["date"].dt.month
date_only = weather.loc[:,["date","month","year"]]
date_only["week"] = date_only["date"].dt.isocalendar().week
date_only.drop_duplicates(inplace=True)

week_closing = date_only.groupby(["year","week"])["date"].max().reset_index()
week_closing = week_closing.rename(columns = {"date":"week_closing"})
week_closing["year-week"] = week_closing["year"].astype("str")+week_closing["week"].astype("str")
week_closing
```

```python
# Create a master reference that maps date to corresponding closing date for the week
date_only["year-week"] = date_only["year"].astype("str")+date_only["week"].astype("str")
date_mapper = date_only.merge(week_closing[["year-week","week_closing"]], left_on="year-week",right_on="year-week",how="left")
date_mapper
```

```python
# Now merge this master reference to trap_by_date dataset so we can do a weekly aggregation 
# of the number of mosquitos caught and the incidence of WNV by trap

trap_by_date = trap_by_date.merge(date_mapper[["date","week_closing"]], left_on="date",right_on="date",how="left")
trap_by_date
```

```python
trap_stats_by_week = trap_by_date.groupby("week_closing")[["nummosquitos","wnvpresent","trap_count"]].sum()
trap_stats_by_week["no_wnv"] = trap_stats_by_week["trap_count"]-trap_stats_by_week["wnvpresent"]
trap_stats_by_week
```

```python
# finally, merge datasets that contain weather data + weekly trap data

weat_trap = weather_subset.merge(trap_stats_by_week,left_index=True,right_index = True, how="left").fillna(0)
for col in ["nummosquitos","wnvpresent","trap_count","no_wnv"]:
    weat_trap[col] = weat_trap[col].astype(int)
weat_trap.head(3)
```

#### Functions to plot bar+line chart combo in subplots:

```python
def weather_plotter(feat,title):
    fig, axes = plt.subplots(nrows=2, ncols=2, sharex=False, 
                                 sharey=True, squeeze=True, figsize=(20,10), constrained_layout = True)
    axes = axes.ravel() 

    for i,year in enumerate(year_list):
        weat = weather_subset.loc[weather_subset["year"]==year,:]
        a = weat_trap.loc[weat_trap["year"]==year,["wnvpresent","no_wnv"]]
        b = weat[feat]
        axes[i].bar(a.index, a["no_wnv"], label='no_wnv', width = 2)
        axes[i].bar(a.index, a["wnvpresent"], bottom=a["no_wnv"],label='wnvpresent', width = 2)
        axes[i].legend(title="Trap Counts:")
        axes[i].set_ylabel("trap_count")
        axes[i] = axes[i].twinx()
        
        if "tavg" in feat:
            c = "red"
        elif "precip" in feat:
            c = "green"
        else:
            c = "purple"
        b.plot(ax = axes[i], color = c)
        
        y_min = b.min()-b.max()
        y_max = b.max()+b.std()
        axes[i].set_ylim(y_min,y_max)


        axes[i].set_title(f'{title} {year}',fontsize=15,fontweight="heavy")
        if "Temp" in title:
            axes[i].set_ylabel(f"\N{DEGREE SIGN}F")
        elif "Precip" in title:
            axes[i].set_ylabel("INCHES")
        else:
            axes[i].set_ylabel("%");
        


    
```

```python
def weather_plotter_by_vol(feat,title):
    fig, axes = plt.subplots(nrows=2, ncols=2, sharex=False, 
                                 sharey=True, squeeze=True, figsize=(20,10), constrained_layout = True)
    axes = axes.ravel() 

    for i,year in enumerate(year_list):
        weat = weather_subset.loc[weather_subset["year"]==year,:]
        a = weat_trap.loc[weat_trap["year"]==year,["nummosquitos"]]
        b = weat[feat]
        axes[i].bar(a.index, a["nummosquitos"], label='nummosquitos',width = 2)
        axes[i].legend()
        axes[i].set_ylabel("nummosquitos")
        axes[i] = axes[i].twinx()
        
        if "tavg" in feat:
            c = "red"
        elif "precip" in feat:
            c = "green"
        else:
            c = "purple"
        b.plot(ax = axes[i], color = c)
        
        y_min = b.min()-b.max()
        y_max = b.max()+b.std()
        axes[i].set_ylim(y_min,y_max)


        axes[i].set_title(f'{title} {year}',fontsize=15,fontweight="heavy")
        if "Temp" in title:
            axes[i].set_ylabel(f"\N{DEGREE SIGN}F")
        elif "Precip" in title:
            axes[i].set_ylabel("INCHES")
        else:
            axes[i].set_ylabel("%");
        

```

```python
year_list = [2007,2009,2011,2013]
weat_feat = ["tavg_rolling_week_average","preciptotal_rolling_week_average","humidity_rolling_week_average"]
weat_title = ["Trap Counts (by incidence of WNV) vs Temp", "Trap Counts (by incidence of WNV) vs TotalPrecip","Trap Counts (by incidence of WNV) vs Humidity"]
weat_label = [f"\N{DEGREE SIGN}F","inches","%"]
```

#### Visualisation: Trap Counts (by incidence of WNV) vs Temperature Features

The model indicated that week is an important predictor of presence of WNV. Indeed, if we zoom into weekly plots of presence of WNV by trap counts, WNV tends to start appearing in the last week of July each year.

Hence, **it is recommended to ramp up vector control measures between mid-Jul to end-Sep each year**. The frequency of application throughout the season will depend on the method of vector control used. For example, some larvicidal agents last for 30 days, compared to the limited efficacy of sprayed adulticides. 

It is important to take note that weather conditions can affect the timing of the appearance of WNV. For example, in 2013, WNV was detected in Week 2 of July instead of the typical Week 4. This is likely due to higher than usual levels of precipitation in the preceding weeks. 

Therefore, if there is **significantly higher levels of precipitation in end June**, we should start vector control in early July instead.

```python
for feat,title in zip(weat_feat,weat_title):
    weather_plotter(feat,title)
```

```python
weat_title_2 = ["Total Mosquitos Trapped vs Temp", "Total Mosquitos Trapped vs TotalPrecip","Total Mosquitos Trapped vs Humidity"]
for feat,title in zip(weat_feat,weat_title_2):
    weather_plotter_by_vol(feat,title)
```

## Target locations for vector control

We should focus on hotspots, ie: locations where WNV is frequently detected. 

```python
## aggregating train dataset to get a summary of traps by coordinates
trap_year_count = train_agg.groupby(["trap","year",'longitude','latitude'])[['nummosquitos','wnvpresent']].sum(['nummosquitos','wnvpresent']).reset_index(drop=False)
trap_year_count['wnvpresent'] = trap_year_count['wnvpresent'].map(lambda x:1 if x>0 else 0)
trap_stats = trap_year_count.groupby(["trap",'longitude','latitude'])[['nummosquitos']].mean(['nummosquitos']).reset_index(drop=False)
trap_stats.set_index("trap",drop= True,inplace=True)
trap_stats.head()
```

```python
trap_year_count["year"].value_counts()
```

```python
wnv_traps = trap_year_count.loc[trap_year_count['wnvpresent']==1]
pos_trap_count = pd.crosstab(wnv_traps["trap"],wnv_traps["year"])
pos_trap_count["frequency"] = pos_trap_count[2013]+pos_trap_count[2011]+pos_trap_count[2009]+pos_trap_count[2007]
pos_trap_count["latest"] = pos_trap_count[2013].map(lambda x: 1 if x==1 else 0)
pos_trap_count.head()
```

```python
#total number of traps with at least one occurence of WMV in 4 years 
pos_trap_count.shape[0]
```

```python
#total number of traps with at least one occurence of WMV in 4 years + recent activity
active_spots = pos_trap_count.loc[pos_trap_count["latest"]==1,
                                   ["frequency"]].value_counts()
active_spots.sort_values()
```

```python
active_spots.sum()
```

**Out of the 138 traps locations, 97 register at least one incidence of WMV between 2007-2013. Out of these 97 spots, we can focus on locations with high frequency of WMV incidence, specifically starting with those with the most recent incidence of infection in the previous year (adding up to 59 counts).** 


```python
all_traps = pd.crosstab(trap_year_count["trap"],trap_year_count["year"])
all_t = all_traps.sum()
pos_t = pos_trap_count.sum()
trap_stats_by_year = pd.DataFrame([all_t,pos_t]).T
trap_stats_by_year.columns = trap_stats_by_year.columns.map({0:"traps_surveyed",1:"positive_traps"})
trap_stats_by_year.drop(labels = ["frequency","latest"], axis = 0, inplace=True)
trap_stats_by_year[["traps_surveyed","positive_traps"]] = trap_stats_by_year[["traps_surveyed","positive_traps"]].astype(int)
trap_stats_by_year["negative_rate"] = trap_stats_by_year["traps_surveyed"]-trap_stats_by_year["positive_traps"]
trap_stats_by_year["positive_rate"] = trap_stats_by_year["positive_traps"]/trap_stats_by_year["traps_surveyed"]
trap_stats_by_year
```

```python
figure, axes = plt.subplots()
trap_stats_by_year.iloc[:,[2,1]].plot(kind="bar",stacked=True,figsize = (8,4),ax = axes),
axes.set_title(label = "Trap Counts by Year", fontsize = 15, fontweight = "heavy");
```

## Conclusion

1. We built 5 classifier models to predict incidence of WMV based on data collected across 4 years, and selected the **XGBoost Classifier** based on its' sensitivity of 85.7%. The model identified the following **features** with high importance:  
   a) The week in which sample of mosquitos was collected  
   b) Weather conditions: Rain, Mist  
   c) Trap cluster   
   d) Species of Mosquito  
   
   
2. We also performed cost-benefit analysis to identify potential strategies to focus vector control efforts, mainly by:  
   a) Identifying seasonal trends for WNV - we propose that **active surveillance and vector control efforts begin in mid-July, subject to weather conditions.** 
      - Warmer temperatures would mean a shorter *Culex* mosquito life cycle, which would mean we need extra cycles of vector control. 
      - Higher than usual precipitation at the end of June would also indicate an early start of the season of breeding and infection, which translates to the need to start vector control in early July.  
        
   
   b) Identifying **clusters of trap locations as hotspots to focus vector control efforts**, whether by increasing frequency of spraying or applying other methods that are more effective.
      - 59 trap locations were identified as hotspots (reported WNV infections in at least 2 years out of the 4 surveyed) or spots with recent activity (reported WNV infection in 2013). 
      - By managing these hotspots more effectively, we will reduce spread of the virus to other locations, minimizing the need for extensive vector control (and therefore, costs) in the long run

### Proposed Future Works:

1. Refine the selected model to increase sensitivity by:  
     a) **Re-enginnering current features**. For example, instead of taking rolling 3-day averages of weather features, we could attempt to see if better results can be obtained with rolling 7-day averages of weather features.     
     b) Adding more features based on existing dataset. Eg: **Wind Speed, Wind Direction and other specific weather conditions (thunderstorms, squall, mist or fog)**  
     c) Collecting and studying additional data for additional features, such as **geographical features, effect of vector control efforts or distribution/density of specific bird species that act as carrier hosts.**   

2. Build and evaluate a **Recurrent Neural Network** model to test if it provides better sensitivity than the current XG Boost model.
